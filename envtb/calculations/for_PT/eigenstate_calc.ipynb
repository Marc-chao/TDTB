```python
def theta(x):
    if x == 0:
        return 0.5

    return 0 if x < 0 else 1

xc = Nx/2
yc = Ny/2
X0 = xc
y0 = 100
k = -1
c = y0*np.sqrt(2)-x0
d = np.sqrt(2)*y0
    
f = (lambda x,y: 100.0*theta(abs(y-yc)-k*(abs(x-xc))-d)*theta(abs((x-xc))-c))
U = envtb.ldos.potential.Potential2DFromFunction(f)
ham_pot = ham.apply_potential(U)
```
